```python
# Importação das bibliotecas

from igraph import Graph
from igraph import plot
```

```python
# Definição do grafo com as arestas
# Use a quantidade de vértices para nomear os vértices


grafo1 = Graph(edges = [(0, 1),(1, 2),(2, 3),(3, 0)], directed = True)

# Definição do rótulo de cada vértice

grafo1.vs['label'] = range(grafo1.vcount())
print(grafo1)


# **Graph()** -> Na primeira linha, estamos criando um objeto grafo1 do tipo Graph. 
# O objeto é inicializado com uma lista de tuplas que representam as arestas do grafo. 
# Cada tupla contém dois números inteiros que representam os vértices conectados pela aresta. 
# No caso, o grafo tem quatro vértices numerados de 0 a 3 e quatro arestas que conectam os vértices em um ciclo. 
# O parâmetro **directed** é definido como True, o que significa que o grafo é direcionado. 
# Obs: **edges** == arestas

# **vs['label']** -> Na segunda linha, estamos adicionando rótulos aos vértices do grafo. 
# O atributo vs representa todos os vértices do grafo e label é um atributo que armazena o rótulo de cada vértice. 
# Estamos definindo os rótulos como uma sequência numérica de 0 a 3.

# **vcount()** ->  método do objeto Graph que retorna o número de vértices no grafo. 

# **IGRAPH D--- 4 4 -- (saída)** -> A linha "IGRAPH D--- 4 4 --" contém a descrição da saída do grafo. 
# Cada termo tem um valor semântico específico, conforme explicado abaixo:

# IGRAPH: É o tipo do objeto de rede, indicando que é um grafo.

# D---: Essa sequência de caracteres representa os atributos do grafo. 
# Em particular, "D" indica que o grafo é direcionado (os nós possuem uma direção definida), 
# enquanto os traços "---" indicam que o grafo não possui outros atributos específicos (como pesos nas arestas ou cores nos nós).

# 4: Esse número representa a quantidade total de nós (vértices) no grafo.

# 4: Esse número representa a quantidade total de arestas no grafo.

# --: O traço duplo final indica que não há mais informações a serem exibidas.
```

```python
# Plotagem do Gráfico


plot(grafo1, bbox = (0, 0, 300, 300))

# **bbox** -> o parâmetro bbox é usado para definir as dimensões da caixa delimitadora (bounding box) que envolve a área
# do gráfico. A caixa delimitadora é usada para definir a área do gráfico que será exibida na tela.
# O parâmetro bbox é uma tupla de quatro valores que representam as coordenadas (x0, y0, x1, y1) da caixa delimitadora. 
# Os valores x0 e y0 representam as coordenadas do canto inferior esquerdo da caixa delimitadora, 
# enquanto os valores x1 e y1 representam as coordenadas do canto superior direito.

# No exemplo, o valor (0, 0, 300, 300) é usado para definir a caixa delimitadora. 
# Isso significa que a área do gráfico será limitada a uma caixa retangular com largura de 300 pixels e altura de 300 pixels, 
# começando no canto inferior esquerdo da tela.
```

```python
# Exemplo dado pelo copiloto

import igraph
from igraph import plot

# Criar o grafo a partir de uma lista de arestas
g = igraph.Graph([(0, 1), (0, 2), (1, 2), (3, 4), (3, 5), (4, 5)])

# Atribuir nomes aos vértices
g.vs["name"] = ["A", "B", "C", "D", "E", "F"]

# Plotar o grafo
plot(g)

# Encontrar os cliques do grafo
cliques = g.cliques()

# Mostrar os cliques encontrados
for clique in cliques:
    # Obter os nomes dos vértices do clique
    names = [g.vs[i]["name"] for i in clique]
    # Mostrar o clique na forma de conjunto
    print(set(names))
```
